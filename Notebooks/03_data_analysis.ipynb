```python
# Import the necessary libraries

import pandas as pd
import numpy as np
```

```python
# Import the data 

df = pd.read_csv("../Data/clean_data/fires_year_state.csv")
```

```python
df
```

```python
# Total fires in brazil per year

total_bra_year = df.groupby("year").sum()[['number']]
total_bra_year
```

```python
# Top 5 years with highest number of cases

total_bra_year.nlargest(5, "number")
```

```python
total_state_year = df.groupby(["year", "month"]).sum()[['number']]
total_state_year
```

```python
# Number of cases per state

state = df.groupby("state").sum()[['number']]
state
```

```python
# Top 5 states with most fires

state.nlargest(5, "number")
```
