```python
# Rename Columns Statistics and Temperature 


temp["Months"]=temp[" Statistics"]

temp=temp.rename(columns={"Temperature - (Celsius)" : "Temperature (Celsius) "})

temp.head()


```

```python
# Remove columns Statistics, ISO3 and Country
temp=temp.drop(columns=[" Statistics", " ISO3", " Country" ])
```

```python
# Remove Average in rows of column Months

#temp['Months'] = temp['Months'].map(lambda x: x.rstrip(' Average'))
#temp

temp['Months'] = temp['Months'].str.replace(" Average","")

temp
```
